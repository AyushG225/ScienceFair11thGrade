```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("llcp2022.csv")
```

```python
data.head()
```

```python
data = data.dropna(axis=1, thresh=400000)
data = data.dropna()
data.shape
```

```python
data.head()
```

```python
data = data[(data.DIABETE4 == 1) | (data.DIABETE4 == 3)]
```

```python
data.shape
```

```python
data.DIABETE4.value_counts()
```

```python
diabetes_columns = ['DIABETE4', 'SEXVAR', 'INCOME3', '_RFHLTH', '_RFDRHV8', '_AGE80', 'SMOKE100', '_METSTAT', '_PRACE2', 'MENTHLTH']
```

```python
data = data.filter(diabetes_columns, axis=1)
```

```python
data.loc[data["DIABETE4"] == 1.0, "DIABETE4"] = 0.0
data.loc[data["DIABETE4"] == 3.0, "DIABETE4"] = 1.0
```

```python
data.to_csv('diabetes_data.csv')
```
